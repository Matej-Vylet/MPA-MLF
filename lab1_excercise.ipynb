```python
def fibonacci(upper_threshold: int) -> list:
    res=[0,1]
    x=0
    y=1
    while y+x<=upper_threshold:
      z=x+y
      x=y
      y=z
      res.append(z)
    return res

print(fibonacci(10))

```

```python
#Rock-Paper-Scissors
def rock_paper_scissors(pl_win, cmp_win):
    import random

    comp_guess = random.choice(["rock", "paper", "scissors"])
    player_input = input("Input your guess: ").lower()
    print("Computer chose",comp_guess)

    if player_input == comp_guess:
      print("Tie")
      return pl_win, cmp_win

    match player_input:
      case "rock":
        if comp_guess == "scissors":
          print("You win")
          pl_win+=1
        else:
          print("You lose")
          cmp_win+=1

      case "paper":
        if comp_guess == "rock":
          print("You win")
          pl_win+=1
        else:
          print("You lose")
          cmp_win+=1

      case "scissors":
        if comp_guess == "paper":
          print("You win")
          pl_win+=1
        else:
          print("You lose")
          cmp_win+=1

      case _:
        print("Invalid input")
    return [pl_win, cmp_win]


num_rounds = 3
pl_win=0
cmp_win=0

for i in range(1,num_rounds+1):
  [pl_win, cmp_win] = rock_paper_scissors(pl_win, cmp_win)
  print("Ended round {}, score: player {}:computer {}".format(num_rounds, pl_win, cmp_win))
```

```python
import numpy as np
import time

def create_array_nxn(n: int) -> np.ndarray:
    if n<=0:
      raise Exception("Invalid dimension n")
    arr=np.arange(n*n-1, -1, -1).reshape(n,n)
    return arr

def apply_threshold_loop(arr: np.ndarray, threshold: int) -> np.ndarray:
    for i in range(len(arr)):
      for j in range(len(arr)):
        if arr[i,j]<threshold:
          arr[i,j]=0
    return arr

def apply_threshold_vectorized(arr: np.ndarray, threshold: int) -> np.ndarray:
    b=arr>=threshold
    return arr*b
    print(b)

def compare_performance(n: int, threshold: int) -> None:
    a = create_array_nxn(n)

    start=time.time()
    b = apply_threshold_loop(a, threshold)
    end = time.time()
    #print(b)
    print("Loop-based approach time: ",end-start)

    start=time.time()
    c = apply_threshold_vectorized(a, threshold)
    end = time.time()
    #print(c)
    print("Vectorized approach time: ",end-start)


compare_performance(100, 50)
```
